```python
import json
import random
import os
import datasets
from tqdm import tqdm
```

```python
def read_query(file):
    dict = {}
    with open(file, 'r') as f:
        for line in f:
            qid, text = line.split('\t')
            dict[qid] = text.strip()
    return dict

def read_qrel(file):
    lst = []
    with open(file, 'r') as f:
        for line in f:
            try:
                qid, _, docid, _ = line.split('\t')
            except:
                qid, _, docid, _ = line.split(' ')
            docid = int(docid)
            lst.append([docid,qid])
    return lst
```

```python
total_num=300000
save_dir='../data/msmarco_data/300k'
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
```

```python
random.seed(313)
data = datasets.load_dataset('Tevatron/msmarco-passage-corpus', cache_dir='../cache')['train']
corpus = [item for item in tqdm(data)]
random.shuffle(corpus)
dev_query = read_query('../data/msmarco_data/dev.query.tsv')
dev_qrel = read_qrel('../data/msmarco_data/qrels.dev.small.tsv')
train_query = read_query('../data/msmarco_data/train.query.tsv')
train_qrel = read_qrel('../data/msmarco_data/qrels.train.tsv')
dl19_query = read_query('../data/msmarco_data/msmarco-19-43-query.tsv')
dl19_qrel = read_qrel('../data/msmarco_data/2019-qrel-pass-rel.txt')
dl20_query = read_query('../data/msmarco_data/msmarco-20-54-queries.tsv')
dl20_qrel = read_qrel('../data/msmarco_data/2020-qrel-pass-final-rel.txt')
```

```python
random.shuffle(dev_qrel)
random.shuffle(train_qrel)
random.shuffle(dl19_qrel)
random.shuffle(dl20_qrel)
print(f'dev_qrel:{len(dev_qrel)},\n train_qrel:{len(train_qrel)},\n dl19_qrel:{len(dl19_qrel)},\n dl20_qrel:{len(dl20_qrel)}')
necessary_list=[item[0] for item in dev_qrel]+[item[0] for item in dl19_qrel]+ [item[0] for item in dl20_qrel]
necessary_doc_set=set(necessary_list)
len(necessary_doc_set)
```

```python
dev_query_num=len(dev_query)+len(dl19_query)+len(dl20_query)
train_query_num=total_num-dev_query_num
train_query_list=[]
train_query_select={}
train_qrel_select=[]
for item in tqdm(train_qrel):
    if item[1] not in train_query_select:
        train_query_list.append(item[1])
        train_qrel_select.append(item)
        train_query_select[item[1]]=train_query[item[1]]
    if len(train_query_list)==train_query_num:
        break
print(f'train_query_select:{len(train_query_select)}, train_query_list:{len(train_query_list)}, train_qrel_select:{len(train_qrel_select)}')
```

```python
train_doc_select=set([item[0] for item in train_qrel_select])
query_doc_num=len(train_doc_select.union(necessary_doc_set))
query_doc_set=train_doc_select.union(necessary_doc_set)
addtional_doc_num=total_num-query_doc_num
query_doc_num
```

```python
#generate id map dict
id_old_to_new={}
def assign_new_id(doc_set):
    for i,old_id in enumerate(query_doc_set):
        id_old_to_new[old_id]=i
    print(len(id_old_to_new))
    return id_old_to_new
id_old_to_new=assign_new_id(query_doc_set)
import pickle
with open(f'{save_dir}/id_old_to_new.pkl', 'wb') as outf:
    pickle.dump(id_old_to_new, outf)
```

```python
#generate doc corpus and doc jsonl
doc_corpus_path=f'{save_dir}/msmarco_300k_doc_collection.tsv'
doc_jsonl_path=f'{save_dir}/msmarco_300k_doc_collection.jsonl'
with open(doc_corpus_path,'w') as doc_corpus, \
     open(doc_jsonl_path,'w') as doc_jsonl:
    for k in id_old_to_new:
        text=data[k]['text']
        new_id=id_old_to_new[k]
        doc_corpus.write(f'{new_id}\t{text}\n')
        jitem=json.dumps({'id':new_id,'contents':text})
        doc_jsonl.write(jitem+'\n')
```

```python
#generate dev, dl19, dl20 queries with label
dev_path=f'{save_dir}/msmarco_dev_data.json'
dl19_path=f'{save_dir}/msmarco_dl19_data.json'
dl20_path=f'{save_dir}/msmarco_dl20_data.json'
output_paths=[dev_path, dl19_path, dl20_path]

def read_qrel_dict(file):
    dict = {}
    with open(file, 'r') as f:
        for line in f:
            try:
                qid, _, docid, _ = line.split('\t')
            except:
                qid, _, docid, _ = line.split(' ')
            docid = int(docid)
            if qid not in dict:
                dict[qid] = [docid]
            else:
                dict[qid].append(docid)
    return dict

dev_qrel = read_qrel_dict('../data/msmarco_data/qrels.dev.small.tsv')
dl19_qrel = read_qrel_dict('../data/msmarco_data/2019-qrel-pass-rel.txt')
dl20_qrel = read_qrel_dict('../data/msmarco_data/2020-qrel-pass-final-rel.txt')
qrels_queries=[[dev_qrel,dev_query], [dl19_qrel,dl19_query], [dl20_qrel,dl20_query]]

for qrel_query, output_path in zip(qrels_queries, output_paths):
    with open(output_path, 'w') as outf:
        for qid in qrel_query[0]:
            q_id = qid
            text = qrel_query[1][qid]
            ori_labels = qrel_query[0][qid]
            new_labels = [id_old_to_new[item] for item in ori_labels]
            jitem = json.dumps({'qid':q_id, 'text': 'Question: '+text, 'text_id': ','.join(map(str, new_labels))})
            outf.write(f'{jitem}\n')
```

```python
#generate train queries with label
train_query_save_path=f'{save_dir}/msmarco_train_query.json'
with open(train_query_save_path, 'w') as outf:
    for qrel in train_qrel_select:
        docid, qid = qrel[0], qrel[1]
        text = train_query_select[qid]
        new_doc_id = id_old_to_new[docid]
        jitem = json.dumps({'qid':qid, 'text':text, 'text_id':new_doc_id})
        outf.write(f'{jitem}\n')
        
```

```python
#generate DSI training data
doc_path=f'{save_dir}/msmarco_300k_doc_collection.jsonl'
doc_data=datasets.load_dataset('json',data_files=doc_path,ignore_verifications=False)['train']
doc_data=doc_data.rename_column('id', 'text_id')
doc_data=doc_data.rename_column('contents','text')
doc_data
```

```python
def add_prompt(example):
    example['text']='Passage: '+example['text']
    return example
doc_data=doc_data.map(add_prompt)
doc_data[0]
```

```python
query_data=datasets.load_dataset('json',data_files=train_query_save_path,ignore_verifications=False)['train']
def add_query_prompt(example):
    example['text']='Question: '+example['text']
    return example
query_data=query_data.map(add_query_prompt)
query_data.to_json(train_query_save_path)
query_data=query_data.remove_columns('qid')
query_data[0]
```

```python
from datasets import concatenate_datasets
# assert query_data.features.type == doc_data.features.type
dsi_train_dataset=concatenate_datasets([doc_data, query_data])
dsi_train_dataset
```

```python
save_path=f'{save_dir}/DSI/msmarco_train_data.json'
# if not os.path.exists(os.path.dirname(save_path)):
#     os.makedirs(save_path)
dsi_train_dataset.to_json(save_path)
```

```python
#generate new qrel
def get_new_qrel(save_path, ori_path):
    with open(ori_path, 'r') as inf,\
         open(save_path, 'w') as outf:
        for line in inf:
            try:
                qid, q0, docid, s = line.split('\t')
            except:
                qid, q0, docid, s = line.split(' ')
            new_id = id_old_to_new[int(docid)]
            outf.write(f'{qid}\t{q0}\t{new_id}\t{s}')
```

```python
save_path=[f'{save_dir}/dev_qrel-newid.txt',
          f'{save_dir}/dl19_qrel-newid.txt',
          f'{save_dir}/dl20_qrel-newid.txt']
ori_path=['../data/msmarco_data/qrels.dev.small.tsv',
         '../data/msmarco_data/2019-qrel-pass-rel.txt',
         '../data/msmarco_data/2020-qrel-pass-final-rel.txt']
for save_p, ori_p in zip(save_path, ori_path):
    get_new_qrel(save_p, ori_p)
```
